```python
arr = dict()
for T in [10, 25]:
    for metric in ['phi_sparsity', 'train_perplexity']:
        for name in ['post_lda', 'post_obd', 'post_obd_limited']:
            with open('20news_experiment/20news_{}t_{}.pkl'.format(T, name), 'r') as f:
                data = pickle.load(f)
            values = np.mean(data[metric], axis=0)
            for it in range(200):
                arr[(T, metric, name, it)] = values[it]
```

```python
lines = []
for T in [10, 25]:
    for name in ['post_lda', 'post_obd', 'post_obd_limited']:
        line = []
        line.append(name)
        line.append(T)
        for metric, signs in [('phi_sparsity', 3), ('train_perplexity', 1)]:
            val_old = round(arr[(T, metric, name, 99)], signs)
            val_new = round(arr[(T, metric, name, 100)], signs)
            line.append(val_old)
            line.append(val_new)
            line.append(int(100 * (val_new - val_old) / val_old))
        print ' & '.join(map(str, line)) + ' \\\\ \\hline'
```

```python
for T in [10, 25]:
    with open('20news_experiment/20news_{}t_plots.pkl'.format(T), 'r') as f:
        data = pickle.load(f)

    font = {
        'family' : 'sans-serif',
        'size'   : 24,
        'sans-serif': ['Times New Roman']
    }
    plt.rc('font', **font)
    plt.figure(figsize=(30, 40))
    col = 0
    for it, n_tw, n_dt, gamma_tw, gamma_dt in data:
        num = col * 100 + it
        plt.subplot(4, 3, num / 20 + 1)
        if it == 80:
            col = 1
        plt.title(u'Итерация №' + str(num))
        plt.xlabel(r'$\log\ n_{wt}$')
        plt.ylabel(r'$\log\ \gamma_{wt}$')
        indices = (n_tw > 1e-5) & (gamma_tw < -1e-5)
        plt.scatter(np.log(n_tw[indices]).flatten(), np.log(- gamma_tw[indices]).flatten())
    plt.tight_layout()
    plt.savefig('sparsity_structure_{}t.png'.format(T), format="png")
    plt.show()
```

```python
from pyartm_experiments.common import default_plot
%matplotlib inline
```

```python
def plot(T, theta_alpha):
    theta_alpha = float(theta_alpha)
    default_plot.eval_experiment_res(
        'plsa', '20news_experiment/20news_{}t_default_{}_{}.pkl'.format(T, 0., 0.),
        'default', '20news_experiment/20news_{}t_default_{}_{}.pkl'.format(T, 0., theta_alpha),
        #'thetaless', '20news_experiment/20news_{}t_thetaless_{}_{}.pkl'.format(T, 0., theta_alpha),
        'transfer', '20news_experiment/20news_{}t_transfer_thetaless_{}_{}.pkl'.format(T, 0., theta_alpha),
    )
```

# T=10

```python
plot(10, 0)
```

```python
plot(10, 0.1)
```

```python
plot(10, 0.1, 0)
```

```python
plot(10, 0, -0.1)
```

```python
plot(10, 0, 0.1)
```

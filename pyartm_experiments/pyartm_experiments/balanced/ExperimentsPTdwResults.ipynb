```python
from pyartm_experiments.common import default_plot
%matplotlib inline
```

```python
def plot(T, phi_alpha, theta_alpha):
    phi_alpha = float(phi_alpha)
    theta_alpha = float(theta_alpha)
    default_plot.eval_experiment_res(
        'default artm', '20news_experiment/20news_{}t_base_{}_{}.pkl'.format(T, phi_alpha, theta_alpha),
        'balanced artm', '20news_experiment/20news_{}t_balanced_ptdw_{}_{}.pkl'.format(T, phi_alpha, theta_alpha),
    )
```

# T=10

```python
plot(10, 0.1, 0.1)
```

```python
plot(10, -0.1, 0)
```

```python
plot(10, 0.1, 0)
```

```python
plot(10, 0, -0.1)
```

```python
plot(10, 0, 0.1)
```
